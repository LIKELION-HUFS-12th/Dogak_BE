```python
import pandas as pd
import os
```

```python
path = '/Users/oesikgogi/Desktop/도서관 정보나루'
files = [f for f in os.listdir(path) if f.endswith('.xlsx')]
```

```python
combined_df = pd.concat([pd.read_excel(os.path.join(path, file)) for file in files], ignore_index=True)
combined_df
```

```python
combined_df.info()
```

```python
combined_df.drop(['번호', '세트 ISBN', '부가기호', '권', '도서권수', '대출건수', '등록일자'], axis=1, inplace=True)
```

```python
# ISBN 컬럼을 기준으로 중복된 행 삭제
combined_df = combined_df.drop_duplicates(subset='ISBN', keep='first')
```

```python
combined_df.reset_index(inplace=True)
```

```python
combined_df.drop('index', axis=1, inplace=True)
```

```python
combined_df.info()
```

```python
combined_df[combined_df['출판사'].isnull()]
```

```python
combined_df = combined_df.dropna(subset=['저자'])
```

```python
# 발행년도와 출판사의 결측치를 0으로 채우기
combined_df['발행년도'] = combined_df['발행년도'].fillna(0)
combined_df['출판사'] = combined_df['출판사'].fillna(0)
```
