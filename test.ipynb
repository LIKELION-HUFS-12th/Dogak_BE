```python
path = '/Users/oesikgogi/Desktop/도서관 정보나루/total 2024-09.csv'
data = pd.read_csv(path)
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
# 숫자로 변환할 수 없는 값은 NaN으로 변환
data['주제분류번호'] = pd.to_numeric(data['주제분류번호'], errors='coerce')

# NaN 값을 -1로 채우고 float 타입 유지
data['주제분류번호'] = data['주제분류번호'].fillna(-1).astype(float)
```

```python
data.info()
```

```python
def classify_subject(code):
    if code == -1:
        return '-'
    str_code = str(int(code))  # 소수점이 있으면 제거하고 문자열로 변환
    if str_code.startswith('0'):
        return '총류'
    elif str_code.startswith('1'):
        return '철학'
    elif str_code.startswith('2'):
        return '종교'
    elif str_code.startswith('3'):
        return '사회과학'
    elif str_code.startswith('4'):
        return '자연과학'
    elif str_code.startswith('5'):
        return '기술과학'
    elif str_code.startswith('6'):
        return '예술'
    elif str_code.startswith('7'):
        return '언어'
    elif str_code.startswith('8'):
        return '문학'
    elif str_code.startswith('9'):
        return '역사'
    else:
        return '기타'

# 새로운 컬럼 추가
data['주제분류'] = data['주제분류번호'].apply(classify_subject)
```

```python
data.to_csv('final 2024-09.csv', index=False)
```
